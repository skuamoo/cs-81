```python
#Load Shakespeare Plays
f = open('shakespeare.txt')
shk = f.read()
shakespeare = eval(shk)
for elem in shakespeare:
    print(elem['title'], "Act:", elem['act'], "Scene:", elem['scene'])
f.close()
```
